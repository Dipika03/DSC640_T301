## Excercise 4.2 Scatterplots, Bubble Charts and Density plot/Maps

```python
## Lets Import the required libraries

import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
## Using crime data and birth rate date for exercise 4.2
## Lets load the data

CrimeRates = pd.read_csv("crimerates-by-state-2005.csv")
BirthRate = pd.read_csv("birth-rate.csv")

## After analysing the data, I noticed District of Columbia data coming as an outlier
## Hence removing the outliers from the data

CrimeRates = CrimeRates[CrimeRates.state != 'District of Columbia']

## Also removing the United States data 

CrimeRates = CrimeRates[CrimeRates.state != 'United States']
```

### Scatterplots

```python
## Lets build the Scatterplot

## Lets initialize the figure variables
f, ax = plt.subplots(figsize=(10, 6))

## Lets work on to show Scatterplot chart
plt.plot('murder', 'burglary', data=CrimeRates, linestyle='none', markersize=10, 
         markerfacecolor="blue", marker='o', markeredgecolor="none")

## Lets add the Scatterplot chart title and labels.
plt.title("Murders vs Burglaries in the United States", fontsize = 18, loc = 'center')
plt.xlabel("Murders", fontsize = 12)
plt.ylabel("Burglaries", fontsize = 12)

plt.show()
```

### Bubble Chart

```python
## Lets build Bubble chart

## Lets initialize the figure variables
f, ax = plt.subplots(figsize=(10, 6))

## Lets work on to show Bubble plot
sc = plt.scatter('murder', 'burglary', s=CrimeRates.population*.0001, data=CrimeRates, 
                 alpha=0.7, color="blue", label='Population')

## Lets add the Bubble chart title and labels.
plt.title("Murders vs Burglaries in the United States", fontsize = 18, loc = 'center')
plt.xlabel("Murders", fontsize = 12)
plt.ylabel("Burglaries", fontsize = 12)

## Lets set the legend for population sizes                   
handles, labels = sc.legend_elements(prop="sizes", alpha=0.4, num=3)     
labels = ["10 MM", "20 MM", "30 MM"]     
legend = ax.legend(handles, labels, loc="lower right", title="Population", 
                   fontsize=11, title_fontsize=14) 

plt.show()
```

### Density Plot

```python
## Lets build Density Plot

## Lets initialize the figure variables
f, ax = plt.subplots(figsize=(10, 6))

## Lets work on to show Density plot
sns.kdeplot(BirthRate['2008'], color="blue", fill="blue")

## Lets add the Density chart title and labels.
plt.title("Global Distribution of Birth Rates in 2008", fontsize = 18, loc = 'center')
plt.xlabel("Live Births per 1,000 Population", fontsize = 12)
plt.ylabel("Density", fontsize = 12)
```
