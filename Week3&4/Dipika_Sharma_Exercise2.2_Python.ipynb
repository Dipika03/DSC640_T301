## Excercise 2.2 Line Charts and Step Charts

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import matplotlib.pyplot as plt
import numpy as np

## Import data for visualization
world_pop = pd.read_excel('world-population.xlsm')


## Let see how the data looks
world_pop.head()
```

### Line Chart

```python
## Lets build Line chart
world_pop.plot(kind='line', x='Year', y='Population', figsize=(10, 6),
          xlabel='Year', ylabel='Population')

## Lets set the title for Line chart
plt.title("World Population by Year", fontsize = 15)
```

### Step Chart

```python
## Lets build step chart
world_pop.plot(kind='line', drawstyle='steps-post', x='Year', y='Population', figsize=(10, 6),
          xlabel='Year', ylabel='Population')

## Setting the tile for step chart
plt.title("World Population by Year", fontsize = 15)
```
